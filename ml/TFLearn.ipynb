```python
import tflearn
import pandas as pd
import numpy as np
from tflearn.layers.core import input_data, dropout, fully_connected  ,reshape
import tensorflow as tf
from tflearn.data_utils import to_categorical
```

```python
ar=np.load('data/rel.npy')
np.random.shuffle(ar)
m=int(ar.shape[0]*0.7)
#traindata=ar[:m,:325]-ar[:m,325:650]
traindata=np.array(ar[:m,:-1],dtype=np.float32)
trainlabel=ar[:m,-1:]
testdata=ar[m:,:-1]
testlabel=ar[m:,-1:]
```

```python
trainlabel2=np.zeros((m,2),dtype=np.float32)
for i in range(trainlabel.shape[0]):
    trainlabel2[i,int(trainlabel[i])]=1
```

```python
net=0
```

```python
net=tflearn.input_data(shape=[None,650])
tflearn.batch_normalization(net)
net=tflearn.fully_connected(net,500)
tflearn.batch_normalization(net)
net=tflearn.fully_connected(net,200)
tflearn.batch_normalization(net)
net=tflearn.fully_connected(net,50)
net=tflearn.fully_connected(net,2,activation='softmax')
net=tflearn.regression(net,loss='categorical_crossentropy')
```

```python
# Define model
model = tflearn.DNN(net)
```

```python
# Start training (apply gradient descent algorithm)
model.fit(traindata, trainlabel2, n_epoch=10, batch_size=10000, show_metric=True)
```
